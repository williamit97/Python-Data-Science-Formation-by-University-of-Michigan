```python
### Si on veut savoir la moyenne des MPG par rapport aux voitures dans notre csv
sum(float(d['cty']) for d in mpg / len(mpg))
```

```python
### Si on veut savoir la moyenne des Highway-autoroute par rapport aux villes dans notre csv
sum(float(d['hwy']) for d in mpg / len(mpg))
```

```python
### Si on veut savoir la moyenne des MPG Miles Per Galon par ville, group by le nombre de cylinders a car has
cylinders = set(d['cyl'] for d in mpg)
cylinders
```

```python
CtyMpgByCyl = []

for c in cylinders:
    summpg = 0
    cyltypecount = 0
    for d in mpg :
        if d['cyl'] == c:
            summpg += float(d['cty'])
            cyltypecount += 1
    CtyMpgByCyl.append((c,summpg / cyltypecount)) #dor ici tuple
    
CtyMpgByCyl.sort(key=lambda x: x[0])
CtyMpgByCyl
```

```python
# On cherche la moyenne Highway MPG pour les différents véhicules
vehicleclass = set(d['class'] for d in mpg)
vehicleclass
```

```python
HwyMpgClass = []

for t in vehicleclass:
    summpg = 0
    vclasscount = 0
    for d in mpg:
        if d['class'] == t:
            summpg += float(d['hwy'])
            vclasscount += 1
    HwyMpgClass.append((c,summpg / cyltypecount)) #dor ici tuple
    
HwyMpgClass.sort(key=lambda x: x[1])
HwyMpgClass           
```

```python
lowercase = 'abcdefghijklmnopqrstuvwxyz'
digits = '0123456789'
correct_answer = []
ID_c = []
ID_d = []

for i in range(0,len(lowercase)):
    for j in range(0,len(lowercase)):
        ID_c.append(lowercase[i]+lowercase[j])
        
for i in range(0,len(digits)):
    for j in range(0,len(digits)):
        ID_d.append(digits[i]+digits[j])
        
for i in range(0,len(ID_c)):
    for j in range(0,len(ID_d)):
        correct_answer.append(ID_c[i]+ID_d[j])
        
answer = [a + b + c + d for a in lowercase for b in lowercase for c in digits for d in digits]

correct_answer == answer
```

# Manipulating Text with Regular Expression

Regular Expression = Regex are written in a condensed formatting langage. Regex = pattern which you give a regex 
processor with some source data. Processor parses the source data using pattern => return a chunck of text back
3 mains utilities : 1) check if a pattern exist within source data 2) get all instances of a complex pattern from source 
data 3) clean source data using pattern through string splitting

```python
# re librairy => store regex
import re
```

## Groups

```python
#We can match different patterns => groups, en même temps. To group patterns => ()

re.findall("([\w ]*)(\[edit\])",wiki) # 1ere paranthese => ts les mots ou espace n'importe quel nb de fois
# 2nd group = balise edition => s'agit headers
```

```python
#Que faire si on veut list de Match object ? => finditer()

for item in re.finditer("([\w ]*)(\[edit\])",wiki):
    print(item.groups()) #return a tuple of group
#On peut obtenir un group => group(nb), group(0) = whole match and the other nb = portion of the match we are interest
```

```python
for item in re.finditer("([\w ]*)(\[edit\])",wiki):
    print(item.group(1))  #Attention ici = item.group pas item.groups
```

```python
for item in re.finditer("([\w ]*)(\[edit\])",wiki):
    print(item.group(0))  
```

```python
#Giving them a label to this group => dictionary. For that => (?P<name>). Parenthese starts the group, ?P => 
#indicate that is an extension to basic regex, <name> = dict key

for item in re.finditer("(?P<title>[\w ]*)(?P<edit_link>\[edit\])",wiki):
    #We can get the dict returned for the item with .groupdict()
    print(item.groupdict()['title'])
```

```python
for item in re.finditer("(?P<title>[\w ]*)(?P<edit_link>\[edit\])",wiki):
    #We can get the dict returned for the item with .groupdict()
    print(item.groupdict()['edit_link'])
```

```python
for item in re.finditer("(?P<title>[\w ]*)(?P<edit_link>\[edit\])",wiki):
    #Afficher les dictionnaires de sorties
    print(item.groupdict())
```

```python
#Il existe d'autres astuces comme \w. a "." for any single character which is not a new line
# \d for any digit, \s for any whitespace charactere => space and tabs
```

## Look-ahead and Look-behind

```python
#Now, si dans mon header j'isoler le texte before the [edit] rendering, sans considération about [edit] text itself.
#Jusque la on a jeté [edit], but now, mais si on veut les utiliser for match, without capture them => 
# mettre dans un groupe "+" use look ahead => "?="

for item in re.finditer("(?P<title>[\w ]+)(?=\[edit\])",wiki):
    #Ce Regex dit de faire correspondre deux groupes, 1er group nommé title
    #le 2eme sera la modification des caract, sans vouloir que cet elem soit mis in my output match objects
    print(item)
```

## Example : Wikipedia Data

```python
with open("buddhist.txt","r") as file:
    wiki = file.read()
wiki
```

```python
#Chaque University suit un meme pattern
#Example => show the verbose mode of python regex. Verbose mode => write multi-line regexes 
#And Increase readability. For this mode => indicate all whitespace charac using \ or \s

pattern="""
(?P<title>.*)          #the university title
(-\ located\ in\ )     #an indicator of the location
(?P<city>\w*)          #city the university is in
(,\ )                  #separator for the state
(?P<state>\w*)         #the state the city is located in"""

#Now we can call finditer() we just pass the re.VERBOSE flag as the last param => more easier to understand large regexes

for item in re.finditer(pattern,wiki,re.VERBOSE):
    #Afficher le dictionnaire
    print(item.groupdict())
```
